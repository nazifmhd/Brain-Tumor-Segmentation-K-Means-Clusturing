# Import Libraries

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
import joblib
```

```python
# Load a image
img = cv2.imread("Datasets/test_images/Tr-gl_0126.jpg", 0)
plt.imshow(img, cmap="gray")
```

```python
# Shape of the image
img.shape
```

```python
512*512
```

```python
height, width = img.shape

# Flatten the image into 1D array
imgFlatten = img.reshape(height*width, 1) 
imgFlatten.shape
```

```python
# Printing a slice of the flattened image
print(imgFlatten[50000:51000])
```

```python
# Model Defining
model = KMeans(n_clusters=5)
```

```python
# Fit the model
model.fit(imgFlatten)
```

```python
# Getting the cluster labels
labels = model.labels_
```

```python
labels.shape
```

```python
labels[50000:51000]
```

```python
# Convert labels to 2D array
labels2D = labels.reshape(height, width)
labels2D.shape
```

```python
# Plot the 2D labels array
plt.imshow(labels2D)
plt.colorbar()
plt.show()
```

```python
# Save the model 
joblib.dump(model, "brain_tumor_kmeans.sav")
```

# Testing The Whole Folder

```python
model = joblib.load("brain_tumor_kmeans.sav")
tumor_label = 3
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# List of image names from the directory
image_names = os.listdir("Datasets/test_images")
print("Images found:", image_names)

# Filter out non-image files 
image_extensions = ['.jpg', '.jpeg', '.png']
image_names = [img_name for img_name in image_names if any(img_name.lower().endswith(ext) for ext in image_extensions)]


for img_name in image_names:
    img_path = os.path.join("Datasets/test_images", img_name)
    
    # Load the original image
    img_original = cv2.imread(img_path)
    if img_original is None:
        print(f"Failed to load image: {img_path}")
        continue  # Skip to the next image if loading fails

    # Convert BGR to RGB for matplotlib
    img_rgb = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)

    img_gray = cv2.imread(img_path, 0)  # Load as grayscale
    if img_gray is None:
        print(f"Failed to load grayscale image: {img_path}")
        continue  # Skip to the next image if loading fails

    height, width = img_gray.shape
    img_flatten = img_gray.reshape(height * width, 1)

    # Predict labels using the model
    labels = model.predict(img_flatten)
    labels2D = labels.reshape(height, width)

    # Create a mask for the tumor
    mask = (labels2D == tumor_label)
    tumor_extracted = np.array((mask * 255), dtype=np.uint8)

    # Erode the tumor mask to refine the contours
    kernel = np.ones((5, 5), np.uint8)
    tumor_extracted = cv2.erode(tumor_extracted, kernel, iterations=1)

    # Find contours in the extracted tumor image
    contours, _ = cv2.findContours(tumor_extracted, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    print("No. Contours:", len(contours))

    # Identify the tumor contour by using OpenCV area
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area > 1000:  # Filter based on area
            cv2.drawContours(img_rgb, [cnt], -1, (0, 255, 0), 3)
            x, y, w, h = cv2.boundingRect(cnt)  # Add the bounding rectangle
            print(f"Bounding Box - x: {x}, y: {y}, w: {w}, h: {h}")
            cv2.rectangle(img_rgb, (x, y), (x + w, y + h), (255, 0, 0), 2)  # Draw rectangle
            cv2.rectangle(img_rgb, (x, y), (x + 120, y - 40), (255, 0, 0), -1)  # Add filled rectangle
            cv2.putText(img_rgb, "TUMOR", (x + 10, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    # Display image using Matplotlib
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.title(f"Processed Image: {img_name}")
    plt.show()

# Clean up any windows opened by OpenCV 
#cv2.destroyAllWindows()
```
